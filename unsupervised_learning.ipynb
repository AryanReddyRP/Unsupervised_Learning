1. KMEANS Clustering

```python
from sklearn.cluster import KMeans
```

```python
k_values = [2,3,4,5,6,7] 
```

```python
ssd = []

for k in k_values:
    km = KMeans(n_clusters=k, max_iter=150, random_state=23)
    km.fit(df_scaled)
    ssd.append(km.inertia_)

ssd
```

```python
plt.plot(ssd)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

k_value = list(range(2,8))
score = []
for k in k_value:
    sc = KMeans(n_clusters=k, random_state=23)
    sc.fit(df_scaled)
    sscore = silhouette_score(df_scaled, sc.labels_)
    score.append(sscore)

score
```

```python
plt.subplot(1,2,1)
plt.plot(k_value, score, marker='o', color='green')
plt.title("silhouette_score vs k")
plt.ylabel("silhouette_score")
plt.xlabel("number of clusters")
plt.xticks(range(2,8))
plt.show()
```

```python
kmodel = KMeans(n_clusters=4, max_iter=150, random_state=23)
kmodel.fit(df_scaled)
```

```python
print(kmodel.labels_)
```

```python
df['cluster_id'] = kmodel.labels_
df.sample(10)
```

```python
group1 = df[df['cluster_id'] == 1]
group1
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Runs', y='Ave', hue='cluster_id', palette='Set1', s=150)
plt.title("This plot will show how cricketers are clustered basedron runs and average")
plt.show()
```

```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=23)
```

```python
x.shape
```

```python
new_df = pd.DataFrame(x)
```

```python
new_df
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.33)
```

```python
from sklearn.cluster import KMeans

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)

wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init="k-means++")
```

```python
y_labels = kmeans.fit_predict(x_train)
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_labels)
plt.show()
```

```python
y_test_labels = kmeans.predict(x_test)
```

```python
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test_labels)
plt.show()
```

```python
from kneed import KneeLocator

kl = KneeLocator(range(1,11), wcss, curve="convex", direction="decreasing")
kl.elbow
```

2. Hierarichal Clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
iris_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
```

```python
x_scaled = ss.fit_transform(iris_data)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(x_scaled)
```

```python
plt.scatter(x_pca[:, 0], x_pca[:, 1])
plt.show()
```

```python
import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20,7))
plt.title("Dendograms")
sc.dendrogram(sc.linkage(x_pca, method='ward'))
plt.title("Dendogram")
plt.xlabel("Sample Index")
plt.ylabel("Eucledian Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, linkage="ward")
cluster.fit(x_pca)
```

```python
cluster.labels_
```

```python
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cluster.labels_)
plt.show()
```

```python
silhouette_coefficients = []

for k in range(2, 11):
    agglo = AgglomerativeClustering(n_clusters=k, linkage='ward')
    agglo.fit(x_pca)
    score = silhouette_score(x_pca, agglo.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of clusters")
plt.ylabel("Sillhoutte Coefficient")
plt.show()
```

3. Clustering Using DBSCAN (Non - Linear Data Distribution)

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
x1, y1 = make_moons(n_samples=250, noise=0.05) # noise is nothing but outliers
```

```python
plt.scatter(x1[:, 0], x1[:, 1])
```

```python
scaler = StandardScaler()
```

```python
x_sc = scaler.fit_transform(x1)
```

```python
# In dbscan we have core point, border point, and outlier points, and we have minimum no of points to decide a particular point belong to which of the above 3 classes and eps value as the radius of circle
dbscan = DBSCAN(eps=0.5)
```

```python
dbscan.fit(x_sc)
```

```python
dbscan.labels_
```

```python
plt.scatter(x1[:,0], x1[:,1],c=dbscan.labels_)
```

```python
plt.scatter(x=x1[:,0], y=x1[:,1],c=y1)
```
